```python
import pandas as pd
import numpy as np
import os


time_period = "14-00--14-10"
date = "20241015"
data_file_name = f"xnas-itch-{date}_{time_period}"
len_chunk = 1
alpha = 0.93
# Prepare the result file for writing
data_file_name = f"prediction_result_{data_file_name}_chunk_size_{len_chunk}_end2end_decay_rate_{alpha}.csv"
data = pd.read_csv(data_file_name)



data["ts_event"] = pd.to_datetime(data["ts_event"])


# reset the index
data = data.reset_index(drop=True)
print("len of selected data", len(data))

```

```python
data["prediction_binary_correctness"] = data["next_price_direction"].astype(int) == data["predicted_direction"].astype(int)
```
