```python
# add synthetic gender column

# Step 1: Get unique userIds
unique_users = data_rating_sorted["userId"].unique()

# Step 2: Randomly assign a gender to each user
user_gender = pd.DataFrame({
    "userId": unique_users,
    "gender": np.random.choice(["M", "F"], size=len(unique_users), p=[0.5, 0.5])
})

# Step 3: Merge the gender assignment back to the original data_rating DataFrame
data_rating_sorted_gender = pd.merge(data_rating_sorted, user_gender, on="userId", how="left")

# Now data_rating_with_gender contains a new 'gender' column

print(data_rating_sorted_gender[1000:1020])
```
